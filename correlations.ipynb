```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import  datetime
import seaborn as sns
import yfinance as yf

sns.set_style("darkgrid")
sns.set(font_scale=1.7)
```

```python
simbols1 = ' HBI   HL   HLX   HRTX   HSC   IMGN   INFN   INN   INO   IOVA   IRWD   JBLU   KOS   KTOS   KYN   LGF-A   LPSN   LUMN   LWLG   LXP   LXRX   MAC   MDXG   MFA   MNKD   MRC   MVIS   NEO   NKTR   NYCB   NYMT   OCSL   OII   OPK   PACB   PBI   PDM   POWW   PSEC   QRTEA   RBBN   REI   RES  RIOT   RWT   SCS   SGMO   SHO   SIRI   SOL'
simbols2 ='    SOL   SRNE   STAR   STKL   SVC   SWN   SXC   TCRT   TELL   TGI   TGTX   TTI   TWO   UAA   UEC   UUUU   VGR   VTNR   VTRS   VUZI   WETF   WKHS   WTI'
tickers_1 = yf.Tickers(simbols1)
simbol_data = tickers_1.history(start = '2010-1-1', actions= False, rounding=True )
```

```python
corr_matrix = simbol_data["Close"].corr().round(2)
positive_corr = corr_matrix > 0.8
negative_corr = corr_matrix < -0.69

plt.figure(figsize=(40,40))
plt.title("Correlation over 0.8")
sns.heatmap(corr_matrix[positive_corr], linewidths=2, linecolor="white", cmap='Greens', annot=True, annot_kws={"size":20}, cbar=False);
```

```python
plt.figure(figsize=(40,40))
plt.title("Correlation below -0.69")
sns.heatmap(corr_matrix[negative_corr], linewidths=2, linecolor="white", cmap='Reds', annot=True, annot_kws={"size":20}, cbar=False);
```
