#### Styling

The plt.plot() function supports many arguments for styling lines and markers:
* color or **c** : set a color of the line
* linestyle or **ls** : choose solid or dashed line
* linewidth or **lw** : line width
* markersize or **ms** : marker size
* markeredgecolor or **mec** : marker's edge color
* markeredgewidth or **mew** : marker's edge width
* markerfacecolor or **mfc** : marker's fill color
* **alpha** : opacity of the plot

To set colors can be used **rgb** codes

The **fmt** argument provides a shorthand for specifying the line style, marker and line color. It can be provided as the third argument to **plt.plot()**

#### fmt = '[marker][line][color]'

If no line style is applied in **fmt**, then only markers will appear

### Bar charts

We might want to draw a bar chart to visualize how the average bill amount varies across different days of the week. One way to do this would be to compute the day-wise averages and then use **plt.bar**.

However, since this is a very common use case, the **Seaborn** library provides a **barplot** function which can automatically compute averages

```python
tips_df = sns.load_dataset("tips")
```

```python
tips_df.groupby('day')[['total_bill']].mean()
```

```python
sns.barplot(data=tips_df, x='day', y='total_bill');
```

```python
sns.barplot(data=tips_df, x='day', y='total_bill', hue='sex');
```

For larger sets of data to display might be vertical bars. All it needs is to change axis information. Which info is less scalable could go on **y** axis.

```python
sns.barplot(data=tips_df, x='total_bill', y='day', hue='sex');
```

### Heatmaps

```python
flights_df = sns.load_dataset("flights").pivot('month', 'year', 'passengers')
```

```python
flights_df
```

```python
plt.figure(figsize=(14,9))
sns.heatmap(flights_df);
```

On heatmap can be seen values of each block as well by using **annot=True**, and use the **cmap** argument to change the color palette

```python
plt.figure(figsize=(14,9))
sns.heatmap(flights_df, fmt='d', annot=True, cmap='Reds');
```

### Plotting multiple charts in a grid

**plt.subplots** returns a set of axes that can be used for plotting

```python
setosa_df = flowers_df[flowers_df.species == "setosa"]
versicolor_df = flowers_df[flowers_df.species == "versicolor"]
virginica_df = flowers_df[flowers_df.species == "virginica"]
```

```python
setosa_df.head(2)
```

```python
fig, axes = plt.subplots(3,3, figsize=(20,15))
axes[0,0].plot(setosa_df.sepal_length, '*--r')
axes[0,0].set_xlabel("times")
axes[0,0].set_ylabel("length")
axes[0,0].set_title("setosa length")
axes[0,1].plot(setosa_df.sepal_width, 'o--r')
axes[0,2].plot(versicolor_df.sepal_width, 'o-b')

axes[1,2].plot(versicolor_df.sepal_length, '-b')
axes[1,0].plot(virginica_df.sepal_width, '-g')
axes[1,1].plot(virginica_df.sepal_length, '-g')

axes[2,0].set_title("scatterPlot")
sns.scatterplot(data=flowers_df, x="sepal_length", y="sepal_width", hue="species", s=100, ax=axes[2,0])
axes[2,1].set_title("Restaurant Bills")
sns.barplot(data=tips_df, x="day", y="total_bill", hue="sex", ax=axes[2,1])
axes[2,2].set_title("Flight traffic")
sns.heatmap(data=flights_df, cmap='Reds', fmt='d', annot=True, ax=axes[2,2])

plt.tight_layout(pad=2) #padding
```
